```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import canny
from skimage import data,morphology
from skimage.color import rgb2gray
import scipy.ndimage as nd
import cv2
from google.colab.patches import cv2_imshow
plt.rcParams["figure.figsize"] = (12,8)
%matplotlib inline
```

```python
image1 = cv2.imread('rambu.jpg')
cv2_imshow(image1)
```

```python
rambu_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(rambu_gray)
```

```python
edges = canny(rambu_gray)
plt.imshow(edges, interpolation='gaussian')
plt.title('Canny detector')
```

```python
from skimage.filters import sobel
elevation_map = sobel(rambu_gray)
fig, ax = plt.subplots(figsize=(4, 3))
plt.imshow(elevation_map, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('on')
ax.set_title('elevation_map')
```
